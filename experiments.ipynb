```python
from graph import draw_graph
from transactions import generate_dataset
import pickle
from algorithms import generate_interesting_patterns, correlation_coefficient, interest_ratio, collective_strength
from metrics import make_experiment
```

#### Генерируем случайный граф или загружаем последние данные.

```python
# PARAMETERS
transaction_count = 50000
count_items = 20
max_count_test = 1000
test_step = max_count_test // 20
replications_count = 50
```

```python
root_of_graph = pickle.load(open('data/last_graph', 'rb'))
# root_of_graph = generate_tree(count_items)
draw_graph(root_of_graph)
print("Graph generated")
```

#### Алгоритм на основе коэффициента корреляции
Генерируем по полученной модели транзакции и находим "истинные" интересные наборы.

```python
# PARAMETERS
threshold = 0.1
parameter = correlation_coefficient
```

```python
dataset = generate_dataset(root_of_graph, transaction_count)
print("Transactions generated")
true_sets_with_value, true_sets = generate_interesting_patterns(dataset, count_items, threshold, parameter)

for true_set in true_sets_with_value:
    print(f'{true_set[0][0]} and {true_set[0][1]} with value = {true_set[1]}')

print("True inresting pattern count:", len(true_sets_with_value))
```

Проводим эксперимент.

```python
make_experiment(max_count_test, test_step, replications_count, root_of_graph, true_sets,
                count_items, threshold, parameter)
```
